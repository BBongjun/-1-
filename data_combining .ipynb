# 행정동코드, 행정동명으로 join

```python
rc= pd.read_csv('행정동별_대규모점포개수.csv',encoding='cp949')
business= pd.read_csv('행정동별_사업체수.csv',encoding='cp949')
hospital= pd.read_csv('행정동별_병원수.csv',encoding='cp949')
food= pd.read_csv('행정동별_음식점수.csv',encoding='cp949')
park= pd.read_csv('행정동별_공원수.csv',encoding='cp949')
cafe= pd.read_csv('행정동별_카페수.csv',encoding='cp949')
house= pd.read_csv('행정동별_주거단지수.csv',encoding='cp949')
bank= pd.read_csv('행정동별_은행.csv',encoding='cp949')
car= pd.read_csv('행정동별_자동차수.csv',encoding='cp949')
school= pd.read_csv('행정동별_학교수.csv',encoding='cp949')
bus_real= pd.read_csv('행정동별_버스_정류장수_승하차인원.csv',encoding='cp949')
subway_real= pd.read_csv('행정동별_지하철역_역사수_승하차인원.csv',encoding='cp949')
people= pd.read_csv('행정동별_연령대_생활인구.csv',encoding='cp949')
```

```python
bike=pd.merge(bike,rc,on='행정동_코드',how='left')
bike=pd.merge(bike,business,on='행정동명',how='left')
bike=pd.merge(bike,hospital,on='행정동_코드',how='left')
bike=pd.merge(bike,food,on='행정동명',how='left')
bike=pd.merge(bike,park,on='행정동_코드',how='left')
bike=pd.merge(bike,cafe,on='행정동_코드',how='left')
bike=pd.merge(bike,house,on="행정동명",how='left')
bike=pd.merge(bike,bank,on='행정동명',how='left')
bike=pd.merge(bike,car,on='행정동명',how='left')
bike=pd.merge(bike,school,on='행정동_코드',how='left')
bike=pd.merge(bike,bus_real,on='행정동_코드',how='left')
bike=pd.merge(bike,subway_real,on='행정동_코드',how='left')
bike=pd.merge(bike,people,on='행정동_코드',how='left')
bike
```

```python
# join하면서 발생한 column 수정
bike=bike.drop(['행정동명_y'],axis=1)
bike=bike.drop(['행정동명'],axis=1)
bike.rename(columns={'행정동명_x':'행정동명'},inplace=True)
bike
```

```python
bike[['남자0~9세','남자10대','남자20대','남자30대','남자40대','남자50대','남자60대','남자70대이상']] = bike[['남자0~9세','남자10대','남자20대','남자30대','남자40대','남자50대','남자60대','남자70대이상']].astype(float)
print(bike['남자70대이상'][1]+bike['남자0~9세'][1])
```

```python
#남/여 전체 생활인구 column 추가하기
bike['남_생활인구']=bike[['남자0~9세','남자10대','남자20대','남자30대','남자40대','남자50대','남자60대','남자70대이상']].sum(axis=1)
bike['여_생활인구']=bike[['여자0~9세','여자10대','여자20대','여자30대','여자40대','여자50대','여자60대','여자70대이상']].sum(axis=1)
bike
```

```python
#결측치는 0으로 바꿔주기
bike=bike.fillna(0)
bike
```
