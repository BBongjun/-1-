# 버스정류장 위치 -> 행정동명,행정동_코드로 변환

```python
#버스정류장 좌표 데이터 불러오기
bus_location = pd.read_csv('서울시버스정류소좌표데이터(20220329).csv',encoding='cp949')
bus_location
```

```python
#좌표 -> 행정동으로 변환
import pandas as pd
import requests 
import sys
import json
import datetime 



def json_request(url='', encoding='utf-8', success=None, error=lambda e: print('%s : %s' % (e, datetime.now()), file=sys.stderr)):
    headers = {'Authorization': 'KakaoAK {}'.format(APP_KEY)}
    resp = requests.get(url, headers=headers)
    # print('%s : success for request [%s]' % (datetime.now(), url))
    return resp.text


def reverse_geocode(longitude, latitude):
    # 파라미터 최적화하여 url 생성
    url = '%s?x=%s&y=%s' %(URL, longitude, latitude)
    # json request
    try:
        # print('try')
        json_req = json_request(url=url)
        json_data = json.loads(json_req)
        json_doc = json_data.get('documents')[1]
        json_name = json_doc.get('region_3depth_name')
        json_code = json_doc.get('code')
    except:
        # print('nan')
        json_name = 'NaN'
        json_code = 'NaN'
    return json_name, json_code

 
def get_address(x,y):
    address = []        
    json_name, json_code = reverse_geocode(x, y) 
    address.append(json_name)
    return address # 전처리 함수에서 주소 리스트 받아서 데이터프레임에 추가


def get_code(x,y):
    code = []        
    json_name, json_code = reverse_geocode(x, y) 
    code.append(json_code)
    return code # 전처리 함수에서 행정구역코드 리스트 받아서 데이터프레임에 추가
    


APP_KEY = 'dc125ad9de42a4ee620f0bdb2e623811' 
URL = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json'

df_all = bus_location
df = df_all[['ARS-ID', 'Y좌표', 'X좌표']] # 기관ID, 병원경도, 병원위도 열만 추출 (도로명주소는 일부러 날림)


for i in range(len(df)): 
    x_crd = float(df.loc[i, ['X좌표']])
    y_crd = float(df.loc[i, ['Y좌표']])
    address = get_address(x_crd, y_crd)
    code = get_code(x_crd, y_crd)
    df.loc[i, ['행정동명']] = address
    df.loc[i, ['행정동_코드']] = code
    
    if i%100 == 0:
        print('{} 번째 경위도 좌표계에 대한 정보 불러오는 중...'.format(i))
    
```

# 버스정류장_행정동코드

```python
#버스정류장_행정동코드 데이터 불러오기
bus_location = pd.read_csv('버스정류장_행정동코드.csv',encoding='cp949')
bus_location
```

```python
#필요한 열만 추출
bus_location=bus_location.filter(['ARS-ID','행정동명','행정동_코드'])
#key로 매칭하기위해 타입 변경
bus_location['ARS-ID']=bus_location['ARS-ID'].astype('str')
bus_location
```

# 행정동별_버스정류장수

```python
#행정동별로 수 세기.
bus_counts=bus_location.groupby(['행정동명','행정동_코드'], as_index=False).count()
bus_counts
```

```python
#칼럼명 수정
bus_counts.rename(columns={'ARS-ID':'정류장수'},inplace=True)
bus_counts
```

# 승하차인원 데이터 전처리

```python
#승하차인원 데이터 불러오기
승하차인원 = pd.read_csv('BUS_STATION_BOARDING_MONTH_202107.csv',encoding='cp949')
승하차인원
```

## 결측치 제거

```python
#결측치 확인했음
승하차인원[승하차인원['버스정류장ARS번호']=='~']
```

```python
#제거
승하차인원 = 승하차인원.drop(index=승하차인원[승하차인원['버스정류장ARS번호']=='~'].index,axis=0)
```

```python
#제거된 모습 확인 가능!
승하차인원[승하차인원['버스정류장ARS번호']=='~']
```

```python
#ARS-ID와 key로 매칭하기위해 타입 변경
승하차인원['버스정류장ARS번호']=승하차인원['버스정류장ARS번호'].astype('int')
승하차인원['버스정류장ARS번호']=승하차인원['버스정류장ARS번호'].astype('str')
#필요한 열만 사용
승하차인원=승하차인원.filter(['버스정류장ARS번호','승차총승객수','하차총승객수'])
승하차인원
```

```python
#행정동명,행정동코드별 승하차수 합 구하기
승하차인원_merged=승하차인원.groupby(['버스정류장ARS번호'],as_index=False).sum()
```

```python
#하루평균 승하차인원
승하차인원_merged['승차총승객수']=round(승하차인원_merged['승차총승객수']/31.0,2)
승하차인원_merged['하차총승객수']=round(승하차인원_merged['하차총승객수']/31.0,2)

승하차인원_merged
```

```python
#정류장위치데이터와 승하차인원데이터 Join
bus_merged = pd.merge(left=bus_location,right=승하차인원_merged,how='inner',left_on='ARS-ID',right_on='버스정류장ARS번호')
bus_merged
```

```python
#승하차승객수를 동별로 합치기
bus_merged=bus_merged.groupby(['행정동명','행정동_코드'], as_index=False).sum()
bus_merged
```

# 승하차인원 + 정류장수 데이터 합치기

```python
#bus_counts 와 bus_merged 합치기
bus_data = pd.merge(left=bus_counts,right=bus_merged,how='inner',on=['행정동명','행정동_코드'])
bus_data
```

```python
#승하차인원 역사별로 보기위해 역사수로 나눔
for i in range(len(bus_data)):
    bus_data['승차총승객수'][i] = bus_data['승차총승객수'][i]/ bus_data['정류장수'][i]
    bus_data['하차총승객수'][i] = bus_data['하차총승객수'][i]/ bus_data['정류장수'][i]
#열이름 수정
bus_data.rename(columns={'승차총승객수':'정류장별_승차승객수','하차총승객수':'정류장별_하차승객수'},inplace=True)
bus_data
```

```python
#csv 내보내기
bus_data.to_csv("행정동별_버스_정류장수_승하차인원.csv",mode="w",encoding='euc-kr',index=False)
```
