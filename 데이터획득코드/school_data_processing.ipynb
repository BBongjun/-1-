# 학교

```python
#학교 데이터 불러오기
school= pd.read_csv('한국교원대학교_초중등학교위치_파일데이터_20210915.csv',encoding='cp949')
school
```

```python
#필요한 열만 추출
school=school[['학교급구분','경도','위도']]
school
```

```python
#학교 경위도 좌표계를 행정동으로
import pandas as pd
import requests 
import sys
import json
import datetime 



def json_request(url='', encoding='utf-8', success=None, error=lambda e: print('%s : %s' % (e, datetime.now()), file=sys.stderr)):
    headers = {'Authorization': 'KakaoAK {}'.format(APP_KEY)}
    resp = requests.get(url, headers=headers)
    # print('%s : success for request [%s]' % (datetime.now(), url))
    return resp.text


def reverse_geocode(longitude, latitude):
    # 파라미터 최적화하여 url 생성
    url = '%s?x=%s&y=%s' %(URL, longitude, latitude)
    # json request
    try:
        # print('try')
        json_req = json_request(url=url)
        json_data = json.loads(json_req)
        json_doc = json_data.get('documents')[1]
        json_name = json_doc.get('region_3depth_name')
        json_code = json_doc.get('code')
    except:
        # print('nan')
        json_name = 'NaN'
        json_code = 'NaN'
    return json_name, json_code

 
def get_address(x,y):
    address = []        
    json_name, json_code = reverse_geocode(x, y) 
    address.append(json_name)
    return address # 전처리 함수에서 주소 리스트 받아서 데이터프레임에 추가


def get_code(x,y):
    code = []        
    json_name, json_code = reverse_geocode(x, y) 
    code.append(json_code)
    return code # 전처리 함수에서 행정구역코드 리스트 받아서 데이터프레임에 추가
    


APP_KEY = 'dc125ad9de42a4ee620f0bdb2e623811' 
URL = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json'

school['행정동명']=''
school['행정동_코드']=''

for i in range(len(school)): 
    x_crd = float(school.loc[i, ['경도']])
    y_crd = float(school.loc[i, ['위도']])
    address = get_address(x_crd, y_crd)
    code = get_code(x_crd, y_crd)
    school.loc[i, ['행정동명']] = address
    school.loc[i, ['행정동_코드']] = code
    
    if i%100 == 0:
        print('{} 번째 경위도 좌표계에 대한 정보 불러오는 중...'.format(i))
    
```

```python
#행정동_코드 8자리로 변경
for i in range(len(school)):
    school['행정동_코드'][i]=school['행정동_코드'][i][0:8]
school
```

```python
#필요한 열만 추출
school=school.filter(['학교급구분','행정동_코드'])
school
```

```python
#행정동별 학교수 구하기
school=school.groupby(school["행정동_코드"],as_index=False).count()
school.rename(columns={'학교급구분':'학교수'},inplace=True)
school
```

```python
#CSV내보내기
school.to_csv("행정동별_학교수.csv",mode="w",encoding='euc-kr',index=False)
```
