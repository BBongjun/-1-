# 자동차 등록현황

```python
#자동차 불러오기
car = pd.read_csv('행정동별 연료별 자동차 등록현황.csv')
car
```

```python
#필요한 열 추출 및 이름변경
car = car.iloc[:,3:]
car = car.dropna(subset=['Unnamed: 3'])
car.rename(columns = {'Unnamed: 3':'주소','Unnamed: 4':'행정동_코드','Unnamed: 5':'용도',
                      'Unnamed: 6':'차종','Unnamed: 7':'연료','Unnamed: 8':'자동차수'}, inplace =True)
car = car.iloc[1:,:]
del car['용도']
del car['차종']
del car['연료']
del car['행정동_코드']
car
```

```python
#feautre type 변경
car = car.astype({'주소':str,'자동차수':int})
car = car[~car["주소"].str.contains('경기도')]
car
```

```python
#주소에서 '행정동' 가져오기
car['행정동명'] = car['주소'].str.split(' ').str[2]
del car['주소']
car
```

```python
#행정동별 자동차수 구하기
cars = car.groupby(car['행정동명'],as_index=False).sum()
cars['행정동명']=cars['행정동명'].astype(str)
cars
```

```python
#CSV내보내기
cars.to_csv("행정동별_자동차수.csv",mode="w",encoding='euc-kr',index=False)
```

# 연령대별 생활인구

```python
#연령대별 생활인구 데이터 가져오기
people = pd.read_csv('LOCAL_PEOPLE_DONG_202107.csv')
people
```

```python
#필요없는 열 삭제
del people['Unnamed: 0']
del people['기준일ID']
del people['시간대구분']
del people['총생활인구수']
```

```python
#행정동코드별 연령대 생활인구수 합 구하기
people2 = people.groupby(['행정동코드'],as_index=False).sum()
people2['행정동코드']=people2['행정동코드'].astype(int)
people2.rename(columns={'행정동코드':'행정동_코드'},inplace=True)
people2
```
